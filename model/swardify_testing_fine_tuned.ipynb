```python
from transformers import pipeline

pipe = pipeline("text-generation", model="bricesuazo/swardify_llama3-8b-instruct-merged_16bit", device = 0)
```

```python
alpaca_prompt = """You are a Swardspeak and Tagalog translator. You only translate input by the user. Do not include any additional information. If you do not know the translation then just return the original input.
### Instruction:
{}

### Input:
{}

### Response:
{}"""
```

```python
def predict(input_text, instruction):
  res = pipe(alpaca_prompt.format(
      instruction,
      input_text,
      ""
  ), max_new_tokens= 128)
  # print(input_text, ":", res[0]["generated_text"].split("\n")[-1])
  return res[0]["generated_text"].split("\n")[-1]
```

```python
predictions_test = []
for data in dataset_test:
  pred  = predict(data["input"], data["instruction"])
  predictions_test.append(pred)

predictions_validation = []
for data in dataset_validation:
  pred = predict(data["input"], data["instruction"])
  predictions_validation.append(pred)
```

```python
from nltk.translate import bleu

def get_each_score(pred, ref):
  scores = []
  for i in range(len(pred)):
    score = bleu(ref[i], pred[i], (1,),)
    scores.append(score)
  return scores
```
